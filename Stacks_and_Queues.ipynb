**1. Parentheses Validator**
>Implement is_valid(s) that checks whether parentheses/brackets/braces are balanced.


```python
'''
I decided to create a function taking in the input from the user as a parameter.
the fuction will check if a bracket was open and, if a corresponding close
bracket was not used, return "Unbalanced". It will obmit any other characters
from this function.

pseudocode code:
define function that takes string as a paramenter:
  declare variable stack as empty list
  initialize variable opens with all opening brackets
  initialize variable closes with all closing brackets

  iterate on each character in the function call's argument:
    if character in open list:
      append to the end of stack list
    else if character in close list:
      if characters not in stack:
        then return False
      initalize variable last with pop() of stack to remove the matched closed character
      if the character in open list index position does not match the character in close:
        then return "Unbalanced"
    else keep going
  once the program has iterated through all characters, check if stack is empty:
    then return "Balanced"
'''
def balanced(expression):
    stack = []
    opens = ["(", "[", "{"]
    closes = [")", "]", "}"]

    for ch in expression:
        if ch in opens:  # compare ch to items in opening bracket list
            stack.append(ch)
        elif ch in closes:  # compare ch to items in closing bracket
            if not stack:
                return False
            last = stack.pop()
            if opens.index(last) != closes.index(ch): # check if indices match
                return 0
        else:
            pass # Ignore non-bracket characters (or return False if strict mode desired)
    if len(stack) == 0:
      return 1
```

```python
# Problem 1 Test
expression = "(]{)"
if balanced(expression):
    print("Balanced")
else:
    print("Unbalanced")
#---------------------------

expression = "([]{})"
if balanced(expression):
    print("Balanced")
else:
    print("Unbalanced")
#---------------------------

expression = "1([0]5{a}b)"
if balanced(expression):
    print("Balanced")
else:
    print("Unbalanced")
```

**2. Postfix Evaluation**
>Evaluate an arithmetic expression given in postfix notation (integers +, -, *, //).

```python
'''
This was an interesting challenge. Since the input is a string literal, when I
cast the string into a list, the integer division would be split into two
conscutive decimal divisions. So I opted to create a loop that checks when a
"/" is followed by another "/", and then replace them both with "//" in the list.

pseudocode Code:
define a postfix function taking one string parameter:
  declare varible stack as empty list (it will store the parsed string)
  initialize variable operators with the arithmatic operations we want to test
  initalize variable tokens by casting the function call's argument as a list
  Since "//" will be parsed as ["/","/"], write a for loop:
    to find two consecutive "/":
      replace the first "/" with "//"
      and delete the second "/"
  iterate on each token in the tokens list:
    initalize variable product with a value of zero (0)
    check if token is equal to any operator in the operators list:
      if not, save the token at the end of the stack list
    else:
      assign the last two entries to variables
        perform the arithmatic operation described by operators:
          and assign it to product
          store the new value in stack list
  return the value at the end of the stack list
'''
def postfix(string):
    stack = []
    operators = ["+", "-", "*", "//"]

    tokens = list(string)

    for i in range(0, len(tokens)-1):
      if tokens[i] == '/' and tokens [i+1] == '/':
        tokens[i] = '//'
        del tokens[i+1]

    for token in tokens:
        product = 0
        if token not in operators:
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()

            if token == "+":
                product = (a + b)
                stack.append(product)
            elif token == "-":
                product = (a - b)
                stack.append(product)
            elif token == "*":
                product = (a * b)
                stack.append(product)
            else:
                product= (a // b)
                stack.append(product)

    return stack
```

```python
# Problem 2 Test
print(postfix('521+4*-3//'))
print(postfix('921+2*-3//'))
```

**3. Queue Using Two Stacks**
>Implement a queue class with enqueue, dequeue, and peek using only two stacks.

```python
'''
Since we are instructed to create a class, all the manupulatons need to be
self contain within the class so they can be inhereted by any other class objects.

pseudocode Code:
Class definition:
  Class constructor:
    two stack variables declared as empty lists

  define enqueue function with two parameters (self as memeber of the class, and x as function argument):
    append x to the end of the stack_in list

  define dequeue variable with only the self parameter as it will not take an argument from the user:
    check if stack_out list is empty:
      while-loop appending the value in stack_in to stack_out:
        then pop the value fron stack_in list
    check if stack_out list is empty:
      raise error if true
    return the value at the end of the stack_out list

    define peek function with only the self parameter as it will not take an argument from the user:
      check if stack_out list is empty:
        while-loop appending the value in stack_in to stack_out:
          then pop the value fron stack_in list
        check if stack_out list is empty:
          raise error if true
    return the value at the end of the stack_out list

    define fucntion to check both lists are empty:
      returns True if enqueue and dequeue lists are empty
'''
class Queue:
    def __init__(self):
        self.stack_in = []   # for enqueue
        self.stack_out = []  # for dequeue and peek

    def enqueue(self, x):
        self.stack_in.append(x) #Add an element to the end of the queue.

    def dequeue(self): # Remove and return the element at the front of the queue.
        if not self.stack_out:
            while self.stack_in: # Transfer all elements from stack_in to stack_out
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            raise IndexError("Dequeue from empty queue")
        return self.stack_out.pop()

    def peek(self): # Return the element at the front without removing it.
        if not self.stack_out:
            while self.stack_in:# Transfer if needed
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            raise IndexError("Peek from empty queue")
        return self.stack_out[-1]

    def is_empty(self):
        return not self.stack_in and not self.stack_out # Check if the queue is empty
```

```python
#Problem 3 Test
q = Queue() # Assign class to variable q
q.enqueue(10) # Store (enqueue) data by making calls to the class enqueue function
q.enqueue(20)
q.enqueue(30)

print(q.peek())
print(q.dequeue())
print(q.dequeue())
print(q.is_empty())
print(q.dequeue())
print(q.is_empty())
```

**4. Min Stack**
>Implement a stack supporting push, pop, top, and get_min (returning min in O(1)).

```python
'''
Just like in the previous problem, since we are instructed to create a class,
all the manupulatons need to be self contain within the class so they can be
inhereted by any other class objects.

My approach was to create two lists (stack and min_stack) where stack will store
all pushed values and min_stack will store the minimum value of the stack. min_stack
is updated to reflect the minimum value of the stack everytime a new value is pushed
to the stack. This means the there is no comparison between the values in stack
whenever the get_min is called, so it runs O(1).
'''
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, numb):
        self.stack.append(numb)
        if not self.min_stack or numb <= self.min_stack[-1]: # If min_stack is empty or numb <= current min, push it
            self.min_stack.append(numb)

    def pop(self):
        if not self.stack:
            raise IndexError("Pop from empty stack")
        val = self.stack.pop()
        if val == self.min_stack[-1]:
            self.min_stack.pop()
        return val

    def top(self):
        if not self.stack:
            raise IndexError("Top from empty stack")
        return self.stack[-1]

    def get_min(self):
        if not self.min_stack:
            raise IndexError("Get_min from empty stack")
        return self.min_stack[-1]

    def is_empty(self):
        return not self.stack
```

```python
s = MinStack()
s.push(5)
s.push(3)
s.push(7)

print(s.get_min())
print(s.pop())
print(s.get_min())
print(s.pop())
print(s.get_min())
```

**5. Sliding Window Maximum**
>Given an array and a window size k, return the maximum in each window.

```python
'''
While reseaching the Real Python site for ideas about how to tackle this problem,
I stumble upon the collections module. I made use of the deque function in the
collections module. This is the website (https://realpython.com/python-collections-module/)

My apprach was to define a function that takes two parameters (a list of values
and the size of the window or the number of items in the window).

pseudocode Code:
define max_window function with two arguments:
  check that second argument is not less than zero:
    return empty list if so
  assigne imported library to variable dq
  declare result variable as empty list

  iterate on each value in the first argument:
    check if indices are outside the window:
      remove values in those indexes
    compare values within the window:
      remove smaller values from back
    store new indexing
    check that index is within range:
      append the value to result list
  return result list
'''

from collections import deque

def max_window(nums, k):
    if not nums or k <= 0:
        return []

    dq = deque()  # stores indices
    result = []

    for i in range(len(nums)):
        # 1
        while dq and dq[0] <= i - k:
            dq.popleft()

        # 2
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        # 3
        dq.append(i)

        # 4
        if i >= k - 1:
            result.append(nums[dq[0]])

    return result
```

```python
print(max_window([1,3,-1,-3,5,3,6,7], 3))
```
